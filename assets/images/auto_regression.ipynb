# Auto Dataset Regression Analysis
This notebook performs simple and multiple linear regression using the Auto dataset.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

df = pd.read_csv('Auto.csv')  # Ensure Auto.csv is in same folder
df.dropna(inplace=True)
df.head()
```

## Simple Linear Regression: MPG ~ Horsepower

```
X = sm.add_constant(df['horsepower'])
y = df['mpg']
model = sm.OLS(y, X).fit()
model.summary()
```

## Multiple Linear Regression: MPG ~ Horsepower + Weight + Acceleration

```
X = df[['horsepower','weight','acceleration']]
X = sm.add_constant(X)
y = df['mpg']
model2 = sm.OLS(y, X).fit()
model2.summary()
```
